```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegressionCV
from sklearn import metrics
from sklearn.model_selection import train_test_split

frame = pd.read_csv('C:\Anaconda\REGRESSION\wcgs.csv', header=0)
```

```python
#creating features df
features = frame.iloc[:,1:9]
features.head()
```

```python
print(features.isna().sum())
print()
print(target.isna().sum())
```

```python
features['chol'].fillna(features.chol.mean(), inplace=True)
```

```python
#encoding categorical data
target=pd.get_dummies(target)
target.head()
```

```python
features=pd.get_dummies(data=features)
features.head()
```

```python
features = features.astype(float)
target = target.astype(float)
```

```python
XTrain, XTest, YTrain, YTest = train_test_split(features, target.chd_yes, test_size=0.2, shuffle=True, random_state=666)
```

```python
logreg = LogisticRegressionCV(cv=5, solver='newton-cg')
logreg.fit(XTrain, YTrain)
# predicting on training and test dataset
predictionTest = logreg.predict(XTest)
predictionTrain = logreg.predict(XTrain)

# calculating efficiencies
effTest = 100 * metrics.accuracy_score(YTest, predictionTest, normalize = True)
effTrain = 100 * metrics.accuracy_score(YTrain, predictionTrain, normalize = True)
print('training set eff =', effTrain, 'test set eff=', effTest)
```
